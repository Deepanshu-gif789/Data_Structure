8. **What is a hash table, and how does it relate to dictionaries in Python?**
  
  - A hash table is a data structure that stores key-value pairs and allows for efficient data retrieval. It uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.
 - **Hash Tables and Python Dictionaries-** - In Python, the dictionary (dict) is implemented using a hash table behind the scenes. When you create a dictionary and assign values to keys, Python performs the following steps internally:
      
     - It hashes the key using a built-in hash function.
  
     - It stores the key-value pair at the corresponding index in a hash table.
    - When you retrieve the value using a key, Python hashes the key again and accesses the value in constant time.

10. **Explain why strings are immutable in Python?**
   -
In Python, strings are immutable, which means that once a string is created, it cannot be changed. Any operation that modifies a string will create a new string object rather than altering the original.

  - **Reasons Why Strings Are Immutable in Python:**
     - **Memory Efficiency and Performance** - Since strings are widely used in programming (e.g., file paths, user inputs, text processing), Python optimizes them by interning frequently used strings. Interning allows multiple variables to point to the same string in memory. This is only safe if strings are immutable—otherwise, changing one would affect all others.
     -  **Hashability** - Immutable objects are hashable, meaning they can be used as keys in dictionaries and elements in sets. Since strings are immutable, they are safe to use as dictionary keys, which would not be possible if they could change.
     - **Safety and Simplicity** - Immutability makes strings safer to use because there’s no risk of accidental modification. It also makes the behavior of strings predictable, which is especially useful in multi-threaded programs where mutable data could lead to conflicts or errors.



11. **What advantages do dictionaries offer over lists for certain tasks?**
   - In Python, both lists and dictionaries are versatile and widely used data structures. However, dictionaries provide significant advantages over lists in situations where key-based access, fast lookups, and data associations are needed.
   - **Advantages of Dictionaries Over Lists:**
       
       -  **Key-Based Access -** Dictionaries store data as key-value pairs, allowing access to data using descriptive keys rather than numeric indices.
       - **Faster Data Retrieval -** Dictionaries use a hash table internally, which allows for constant-time (O(1)) access to values by their keys, regardless of the dictionary’s size. In contrast, lists require a linear-time (O(n)) search to find an element unless you know its exact index.
       -  **Better for Representing Structured Data -** Dictionaries are ideal for modeling real-world entities where data is naturally associated with labels or attributes.

17. **Describe the time complexity of accessing elements in a dictionary.**
    - Accessing elements in a Python dictionary is generally very fast, and here’s why:

     - **Time Complexity:**
        - O(1) on average

        - O(n) in the worst case (but very rare)

        **O(1) on Average?** -
           
           -  Python dictionaries are implemented as hash tables.
           - When you do something like my_dict["key"], Python:
               
               -  Computes the hash of "key".
               - Uses that hash to jump directly to a memory location.
               - Checks for a matching key (using ==).
               - Returns the value if found.
           - This constant-time behavior holds regardless of the size of the dictionary.

      - **O(n) in the worst case (but very rare)** -
          
          - In rare cases—usually due to hash collisions (when multiple keys hash to the same bucket)—you might get O(n) behavior. Python handles this with collision resolution strategies, so it’s generally not a concern unless:
            
            -  You're dealing with malicious input (like in security-sensitive systems)
            - You override __hash__ or __eq__ in a custom object in a weird way

19. **Why are dictionaries considered unordered, and how does that affect data retrieval?**
      
      -  Dictionaries in Python are considered unordered because they use a hash table to store data. This means that the elements are not stored in a specific order based on their position, like in a list. Instead, each value is stored using a unique key, and the key is used to quickly find the value. In versions of Python before 3.7, dictionaries did not preserve the order in which items were added. Starting from Python 3.7, dictionaries now maintain insertion order, but they are still called unordered because data is accessed by key, not by position. This affects data retrieval in that:
       
        -  You cannot access values by index (like dict[0] won't work).
        -  You must always use the key to get the corresponding value (dict["name"]).
